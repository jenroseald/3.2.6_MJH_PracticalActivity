## Q1 What is the average rating per film

```python
# Define user function to find mean of the two columns
```

```python
# Formula is ((IMDb/10) + Rotten Tomatoes) / 2
```

```python
def av_2cols(colm1, colm2):
    """work out average of two columns"""
    colm_av = (colm1/10 + colm2)/2
    return colm_av
```

```python
# Apply function
```

```python
mov_ott_ratings_final = av_2cols(mov_ott_ratings['IMDb'], mov_ott_ratings['Rotten Tomatoes'])
```

```python
mov_ott_ratings_final
```
